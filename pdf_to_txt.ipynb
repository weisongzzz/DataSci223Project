```python
# This script takes the pdf and extracts texts to a new text file

import pdfplumber

# Mannually changed the pdf_path and txt_path variables to write txt files

# Path to the PDF file
pdf_path = 'N5R.pdf'

# Open the PDF file
with pdfplumber.open(pdf_path) as pdf:
    text = ''
    # Extract text from each page of the PDF
    for page in pdf.pages:
        text += page.extract_text()

print(text)

txt_path = 'N5R.txt'
with open(txt_path, 'w', encoding='utf-8') as file:
    file.write(text)
```

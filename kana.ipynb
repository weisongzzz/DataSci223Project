```python
# kana

# This script train the user on hiragana and katakana with English pronounciation
# The user can run either block of code at a time and input answer

# By Weisong Zhang, 3/1/2024
import random
```

```python
hiragana_dict = {
    'あ': 'a',
    'い': 'i',
    'う': 'u',
    'え': 'e',
    'お': 'o',
    'か': 'ka',
    'き': 'ki',
    'く': 'ku',
    'け': 'ke',
    'こ': 'ko',
    'さ': 'sa',
    'し': 'shi',
    'す': 'su',
    'せ': 'se',
    'そ': 'so',
    'た': 'ta',
    'ち': 'chi',
    'つ': 'tsu',
    'て': 'te',
    'と': 'to',
    'な': 'na',
    'に': 'ni',
    'ぬ': 'nu',
    'ね': 'ne',
    'の': 'no',
    'は': 'ha',
    'ひ': 'hi',
    'ふ': 'fu',
    'へ': 'he',
    'ほ': 'ho',
    'ま': 'ma',
    'み': 'mi',
    'む': 'mu',
    'め': 'me',
    'も': 'mo',
    'や': 'ya',
    'ゆ': 'yu',
    'よ': 'yo',
    'ら': 'ra',
    'り': 'ri',
    'る': 'ru',
    'れ': 're',
    'ろ': 'ro',
    'わ': 'wa',
    'を': 'wo',
    'ん': 'n'
}

def quiz_hiragana():
    random_pair = random.choice(list(hiragana_dict.items()))
    question = random_pair[0]
    correct_answer = random_pair[1]

    user_answer = input(f"What is the Hiragana for '{question}'? ").strip()

    if user_answer == correct_answer:
        print("Correct! Well done.")
    else:
        print(f"Oops! The correct answer is '{correct_answer}'.")

if __name__ == "__main__":
    print("Welcome to the Hiragana Quiz!")

    while True:
        quiz_hiragana()
        play_again = input("Do you want to play again? (yes/no) ").lower()
        if play_again != 'yes':
            print("Thanks for playing. Sayonara!")
            break
```
